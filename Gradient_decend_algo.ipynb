```python
import matplotlib.pyplot as plt
x=2
lr=0.01
precision=0.000001
previous_step_size=1
max_iter=10000
iters=0

#gf=lambda x : (x+3)**2
gf = lambda x : 2 * (x + 3)
```

```python
gd = []
```

```python
while precision < previous_step_size and iters < max_iter:
    prev = x
    x = x- lr * gf(prev)
    previous_step_size = abs(x - prev)
    iters+=1
    print('iteractions : ',iters,' Value : ',x)
    gd.append(x)
```

```python
print('local minima : ',x)
```

```python
plt.plot(gd)
plt.show()
```
